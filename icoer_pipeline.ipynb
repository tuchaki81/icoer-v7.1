# ICOER v7.1 — Pipeline de Demonstração
Este notebook demonstra o uso do índice de Coerência Informacional (ICOER) versão 7.1, combinando dados simbólicos e biológicos simulados.

```python
# Imports necessários
from src.icoer_calc import compute_icoer

# Caminhos para dados de exemplo
text_path = 'data/sample_texts.json'
eeg_path = 'data/simulated_eeg.csv'
```

```python
# Executar o cálculo do ICOER
result = compute_icoer(text_path, eeg_path)
result
```

```python
# Visualização simples dos componentes
import matplotlib.pyplot as plt

labels = ['S', 'B', 'T', 'H', 'ICOER']
values = [result[k] for k in labels]

plt.figure(figsize=(8,4))
plt.bar(labels, values, color='skyblue')
plt.title('Componentes do Índice ICOER')
plt.ylabel('Valor Normalizado')
plt.grid(True, axis='y', linestyle='--')
plt.show()
```
